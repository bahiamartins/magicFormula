```python
import yahooquery as yf
```

```python
simbol = 'CMIG4.SA'
```

```python
tkr = yf.Ticker(simbol)
```

```python
tkr.all_modules[simbol]
```

```python
tkr_all = tkr.all_modules[simbol]
```

```python
ebit = tkr_all['incomeStatementHistoryQuarterly']['incomeStatementHistory'][0]['ebit']
ebit
```

```python
#Return on Invested Capital
#ROIC = Operating Income (1 - Tax Rate) / Invested Capital

netIncome_year = tkr_all['incomeStatementHistory']['incomeStatementHistory'][0]['netIncome']
netIncome_year
```

```python
interestExpense_year = tkr_all['incomeStatementHistory']['incomeStatementHistory'][0]['interestExpense']
interestExpense_year
```

```python
incomeTaxExpense_year = tkr_all['incomeStatementHistory']['incomeStatementHistory'][0]['incomeTaxExpense']
incomeTaxExpense_year
```

```python
incomeBeforeTax_year = tkr_all['incomeStatementHistory']['incomeStatementHistory'][0]['incomeBeforeTax']
incomeBeforeTax_year
```

```python
taxRate = incomeTaxExpense_year / incomeBeforeTax_year
round(taxRate*100,2)
```

```python
#Operating Income
opIn = netIncome_year + interestExpense_year * (1-taxRate)
round(opIn)
```

```python
operatingIncome_year = tkr_all['incomeStatementHistory']['incomeStatementHistory'][0]['operatingIncome']
operatingIncome_year
```

```python
# Invested Capital (IC) = Short-term debt + Long-term debt + Shareholder equity - Cash/equivalents - Goodwill

longTermDebt_year = tkr_all['balanceSheetHistory']['balanceSheetStatements'][0]['longTermDebt']
longTermDebt_year
```

```python
shortTermDebt_year = tkr_all['balanceSheetHistory']['balanceSheetStatements'][0]['shortLongTermDebt']
shortTermDebt_year
```

```python
# shareholder equity
totalStockholderEquity_year = tkr_all['balanceSheetHistory']['balanceSheetStatements'][0]['totalStockholderEquity']
totalStockholderEquity_year
```

```python
goodWill_year = tkr_all['balanceSheetHistory']['balanceSheetStatements'][0]['goodWill']
goodWill_year
```

```python
cash_year = tkr_all['balanceSheetHistory']['balanceSheetStatements'][0]['cash']
cash_year
```

```python
totalCashflowsFromInvestingActivities = tkr_all['cashflowStatementHistory']['cashflowStatements'][0]['totalCashflowsFromInvestingActivities']
totalCashflowsFromInvestingActivities
```

```python
totalCurrentLiabilities = tkr_all['balanceSheetHistory']['balanceSheetStatements'][0]['totalCurrentLiabilities']
totalCurrentLiabilities
```

```python
#invested Capital
invCap = totalCurrentLiabilities +longTermDebt_year + shortTermDebt_year + totalStockholderEquity_year - goodWill_year - cash_year - totalCashflowsFromInvestingActivities
invCap
```

```python
ROIC = opIn * (1 - taxRate) / invCap
ROIC
```
